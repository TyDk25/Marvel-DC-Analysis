## AN ANALYSIS ON WHETHER MARVEL OR DC MOVIES PERFORM BETTER

# Plotting with Python always starts with importing the necessary libraries. 

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

# Reading the Dataset

```python
df=pd.read_csv("db.csv", encoding='latin1')
df
```

# Section 1: Cleaning The Data

```python
del df['Unnamed: 0']
```

## As we have called the datatset 'df' we will always refer back to this when accessing the columns within the data.

```python
df["Original Title"].replace("Thor:Ragnarok", "Thor: Ragnarok", inplace=True)
df["Original Title"].replace("Captain Marve", "Captain Marvel", inplace=True)
df["Original Title"].replace("Jonah Hex 	", "Jonah Hex", inplace=True)
```

```python
df['Budget'] = pd.to_numeric(df['Budget'], errors='coerce')
```

```python
df
```

```python
 df.columns.values[6] = 'Budget'
```

## iloc = index locator

```python
df.iloc[3, df.columns.get_loc('Budget')] = 150000000.0
df.iloc[6, df.columns.get_loc('Budget')] = 200000000.0
df.iloc[15, df.columns.get_loc('Budget')] = 175000000.0
df.iloc[18, df.columns.get_loc('Budget')] = 321000000.0
df.iloc[20,df.columns.get_loc('Budget')] = 175000000.0
df.iloc[38,df.columns.get_loc('Budget')] = 55000000.0
df.iloc[28,df.columns.get_loc('Budget')] = 47000000.0
```

```python
df
```

# # Remove whitespace from column names to have easier accessibility

```python
df['Release'] = df['Release'].astype(str).str.replace(r'\.0', '')
```

# # Converting Columns

```python
df.info()
```

```python
df['Release'] = pd.to_numeric(df['Release'])
df['Minutes'] = pd.to_numeric(df['Minutes'], errors='coerce')
df['Rate'] = pd.to_numeric(df['Rate'])
```

```python
pd.options.display.float_format = '{:.0f}'.format
```

# Section 2: Adding Custom Columns and Calculated Fields. 

```python
df['Profit Worldwide'] = df['Gross Worldwide'] - df['Budget']
```

## LOC = Another way to locate columns, used to locate strings.

```python

rowindex = df.index[[2,7,8,16,19,22,26,14,18,21,10]]

df.loc[rowindex, 'Sequels'] = "Sequel"

df["Sequels"] 

groupindex=df.index[[5,9,10,14,18,21,32,35]]

df.loc[groupindex,'Grouped Movies'] = "Grouped"

df["Sequels"].fillna("First Movie", inplace=True)

df["Grouped Movies"].fillna("Not a Group", inplace=True)
```

```python
Marvel = df[df["Company"].str.contains("Marvel")]

DC = df[df["Company"].str.contains("DC")]
```

```python
df.sort_values(by="Profit Worldwide", ascending = False, inplace=True)
```

# SECTION 3 - Plotting :)

```python
top5 = df.sort_values('Rate', ascending=False).head(5)

plt.style.use('fivethirtyeight')


fig, ax = plt.subplots(figsize=(15,9))

ax.set_title('Top 5 Highest Rated Movies Based On IMDB Scores', pad=20)
ax.bar(top5['Original Title'], top5['Rate'], color='green')

ax.set_xlabel('Original Title',fontsize=15,labelpad=20)

ax.set_ylabel('IMDB Ratings',fontsize=15, labelpad=10)
ax.set_yticks([0,2,4,6,8,10])
plt.xticks(rotation = 3,ha='center')

for i, rate in enumerate(top5['Rate']):
    ax.text(i, rate + 0.2, str(rate), ha='center', fontsize=15)
```

```python
fig1, ax1 = plt.subplots(figsize=(10,6))

plt.style.use('ggplot')

ax1.set_title('Does Budget Equal Success in Marvel Movies?', pad=20)

scatter=ax1.scatter(Marvel['Original Title'],Marvel['Rate'], c=Marvel['Budget'],cmap='cool',s=100)

cbar = plt.colorbar(scatter)
cbar.set_label("Budget (M)")

plt.xticks(rotation=50, ha='right', fontsize=10)

ax1.set_ylabel('IMDB Ratings', fontsize=12, labelpad=20)
ax1.set_xlabel('Movie Titles', fontsize=15, labelpad=10)

ax1.tick_params(axis='x', pad=20)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,5))

plt.style.use('fivethirtyeight')

ax.set_title('Does Budget Equal Success in DC Movies?', pad=20)

scatter=ax.scatter(DC['Original Title'],DC['Rate'], c=DC['Budget'],cmap='viridis',s=100)

cbar = plt.colorbar(scatter)
cbar.set_label("Budget (M)", fontsize=12)

plt.xticks(rotation=45, ha='right')

ax.tick_params(axis='x', pad=0)
plt.subplots_adjust(bottom=0.1)


ax.set_ylabel('IMDB Ratings', fontsize=12, labelpad=20)
ax.set_xlabel('Movie Titles', fontsize=12, labelpad=5)

plt.figure()

plt.show()
```

```python
MCUmean=Marvel['Metascore'].mean().round()

MCUmean

DCmean=DC['Metascore'].mean().round()

DCmean
```

```python
plt.figure(figsize=(8,8))

plt.title('Average Metascore Ratings For Marvel & DC Movies', pad= 20)

plt.bar('DC',DCmean, width=0.5, label='DC')
plt.bar('Marvel',MCUmean, width=0.5, label='Marvel')


plt.ylabel('Metascore', fontsize=14, labelpad=10)
plt.xlabel('Company', fontsize=14, labelpad=10)
plt.legend()

plt.text('DC', DCmean + 1, str(DCmean), ha='center', fontsize=13)
plt.text('Marvel', MCUmean + 1, str(MCUmean), ha='center', fontsize=13)

plt.show()
```

```python
data = df.set_index('Original Title')
```

```python
FirstMovie = data[data['Sequels'].str.contains("First Movie")]

Sequel = data[data['Sequels'].str.contains("Sequel")]
```

```python
mean_firstmovie = FirstMovie["Metascore"].mean()

rounded_firstmovie = mean_firstmovie.round()
```

```python
mean_sequel = Sequel["Metascore"].mean()

rounded_sequel = mean_sequel.round()
```

```python
plt.figure(figsize=(10,6))

plt.title('Metascore Ratings Comparison: First Movies vs. Sequels', pad=20)

plt.style.use('ggplot')

plt.bar('First Movies', rounded_firstmovie, width=0.5, label='First Movies')

plt.bar('Sequels', rounded_sequel, width=0.5, label='Sequels')

plt.ylabel('Metascore', fontsize=15, labelpad=10)

plt.text('First Movies',rounded_firstmovie + -5, str(rounded_firstmovie), ha='center', fontsize=13)
plt.text('Sequels',rounded_sequel + -5, str(rounded_sequel), ha='center', fontsize=13)

plt.legend()

plt.show()
```

```python
grouped_movies=data[data["Grouped Movies"].str.contains("Grouped")]

ungrouped_movies=data[data["Grouped Movies"].str.contains("Not a Group")]
```

```python
mean_grouped = grouped_movies["Rate"].mean()

mean_ungrouped = ungrouped_movies["Rate"].mean()

rounded_grouped = mean_grouped.round()

rounded_ungrouped = mean_ungrouped.round()
```

```python
plt.style.use('fivethirtyeight')

plt.title(' IMDB Comparison: Group vs. Solo Movies', pad=20)

plt.bar('Grouped Movies', rounded_grouped, label='Grouped')
plt.bar('Solo Movies', rounded_ungrouped, label='Solo')

plt.ylabel('IMDB Ratings', labelpad=15)

plt.text('Grouped Movies', rounded_grouped + -5, str(rounded_grouped), ha='center', fontsize=16, color='white')
plt.text('Solo Movies', rounded_ungrouped + -5, str(rounded_ungrouped), ha='center', fontsize=16, color='white')

plt.legend()
plt.show()


```

```python
plt.figure(figsize=(14,6))
plt.suptitle('Marvel Movies: Worldwide Profit vs Budget', fontsize=16, y=1.05)

plt.bar(Marvel['Original Title'], Marvel['Profit Worldwide'], label='Profit')
plt.bar(Marvel['Original Title'], Marvel['Budget'],label='Budget')

plt.xlabel('Movie Titles')
plt.ylabel('Profit Made (in billions)')

plt.xticks(rotation=50, ha='right')

plt.legend()



plt.figure(figsize=(14,6))

plt.suptitle('DC Movies: Worldwide Profit vs Budget', fontsize=16, y=1.05)

plt.bar(DC['Original Title'], DC['Profit Worldwide'], color='Gray', label='Profit')
plt.bar(DC['Original Title'], DC['Budget'], color='Red', label='Budget')

plt.ylabel('Profit Made (in billions)')
plt.xlabel('Movie Titles')

plt.legend()

plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
DCProf=DC['Profit Worldwide'].sum()

DCProf_round = round(DCProf / 1e9, 1)
```

```python
MarvelProf=Marvel['Profit Worldwide'].sum()

MarvelProf_round = round(MarvelProf / 1e9, 1)
```

```python
plt.figure(figsize=(11,7))

plt.title('Gross Worldwide Profits: Marvel Vs. DC', pad=10)
plt.style.use('fivethirtyeight')

plt.bar('Marvel',MarvelProf_round, label='Marvel', width=0.2)
plt.bar('DC',DCProf_round, label='DC', width=0.2)

plt.ylabel('Gross Profits Worldwide',labelpad=20)
plt.xlabel('Companies', labelpad=5)

plt.legend()


plt.text('Marvel',MarvelProf_round + 1, f'${MarvelProf_round} (B)', ha='center', fontsize=16, color='black', va='bottom')
plt.text('DC',DCProf_round + 1, f'${DCProf_round} (B)', ha='center', fontsize=16, color='black', va='bottom')

plt.show()
```
