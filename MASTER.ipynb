```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("db.csv", encoding='latin1')
df
```

```python
del df['Unnamed: 0']
```

```python
df["Original Title"].replace("Thor:Ragnarok", "Thor: Ragnarok", inplace=True)
df["Original Title"].replace("Captain Marve", "Captain Marvel", inplace=True)
df["Original Title"].replace("Jonah Hex 	", "Jonah Hex", inplace=True)
```

```python
df['Budget'] = pd.to_numeric(df['Budget'], errors='coerce')
```

```python
 df.columns.values[6] = 'Budget'
```

```python
df.iloc[3, df.columns.get_loc('Budget')] = 150000000.0
df.iloc[6, df.columns.get_loc('Budget')] = 200000000.0
df.iloc[15, df.columns.get_loc('Budget')] = 175000000.0
df.iloc[18, df.columns.get_loc('Budget')] = 321000000.0
df.iloc[20,df.columns.get_loc('Budget')] = 175000000.0
df.iloc[38,df.columns.get_loc('Budget')] = 55000000.0
df.iloc[28,df.columns.get_loc('Budget')] = 47000000.0
```

```python
df
```

```python
df['Release'] = df['Release'].astype(str).str.replace(r'\.0', '')
```

```python
df['Release'] = pd.to_numeric(df['Release'])
df['Minutes'] = pd.to_numeric(df['Minutes'], errors='coerce')
df['Rate'] = pd.to_numeric(df['Rate'])
```

```python
pd.options.display.float_format = '{:.0f}'.format
```

```python
df
```

```python
df['Profit Worldwide'] = df['Gross Worldwide'] - df['Budget']
```

```python

rowindex=df.index[[2,7,8,16,19,22,26,14,18,21,10]]

df.loc[rowindex, 'Sequels'] = "Sequel"

df["Sequels"] 

groupindex=df.index[[5,9,10,14,18,21,32,35]]

df.loc[groupindex,'Grouped Movies'] = "Grouped"

df["Sequels"].fillna("First Movie", inplace=True)

df["Grouped Movies"].fillna("Not a Group", inplace=True)
```

```python
Marvel = df[df["Company"].str.contains("Marvel")]

DC = df[df["Company"].str.contains("DC")]
```

```python
df.sort_values(by="Profit Worldwide", ascending = False, inplace=True)
```

```python
data = df.set_index('Original Title')
```

```python
FirstMovie = data[data['Sequels'].str.contains("First Movie")]

Sequel = data[data['Sequels'].str.contains("Sequel")]
```

```python
mean_firstmovie = FirstMovie["Metascore"].mean()

rounded_firstmovie = mean_firstmovie.round()
```

```python
mean_sequel = Sequel["Metascore"].mean()

rounded_sequel = mean_sequel.round()
```

```python
plt.figure(figsize=(10,6))

plt.title('Metascore Ratings Comparison: First Movies vs. Sequels', pad=20)

plt.style.use('ggplot')

plt.bar('First Movies', rounded_firstmovie, width=0.5, label='First Movies')

plt.bar('Sequels', rounded_sequel, width=0.5, label='Sequels')

plt.ylabel('Metascore', fontsize=15, labelpad=10)

plt.text('First Movies',rounded_firstmovie + -5, str(rounded_firstmovie), ha='center', fontsize=13)
plt.text('Sequels',rounded_sequel + -5, str(rounded_sequel), ha='center', fontsize=13)

plt.legend()

plt.show()
```

```python
grouped_movies=data[data["Grouped Movies"].str.contains("Grouped")]

ungrouped_movies=data[data["Grouped Movies"].str.contains("Not a Group")]
```

```python
mean_grouped = grouped_movies["Rate"].mean()

mean_ungrouped = ungrouped_movies["Rate"].mean()

rounded_grouped = mean_grouped.round()

rounded_ungrouped = mean_ungrouped.round()
```

```python
plt.style.use('fivethirtyeight')

plt.title(' IMDB Comparison: Group vs. Solo Movies', pad=20)

plt.bar('Grouped Movies', rounded_grouped, label='Grouped')
plt.bar('Solo Movies', rounded_ungrouped, label='Solo')

plt.ylabel('IMDB Ratings', labelpad=15)

plt.text('Grouped Movies', rounded_grouped + -5, str(rounded_grouped), ha='center', fontsize=16, color='white')
plt.text('Solo Movies', rounded_ungrouped + -5, str(rounded_ungrouped), ha='center', fontsize=16, color='white')

plt.legend()
plt.show()


```
